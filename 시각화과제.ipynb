**시각화에 이용할 변수를 영어로 변경**  
**(추후 시각화할 때 한글 깨짐 방지하기 위해 영어로 변경)**

```python
health.rename(columns={'시도코드':'City Name'}, inplace=True)
health.rename(columns={'연령대코드(5세단위)':'Age Name'}, inplace=True)
health.rename(columns={'신장(5Cm단위)':'height'}, inplace=True)
health.rename(columns={'체중(5Kg 단위)':'weight'}, inplace=True)
health.rename(columns={'수축기혈압':'systolic blood pressure'}, inplace=True)
health.rename(columns={'이완기혈압':'diastolic blood pressure'}, inplace=True)
health.rename(columns={'치아우식증유무':'cavity'}, inplace=True)
health.rename(columns={'흡연상태':'smoking status'}, inplace=True)
display(health)
```

**LDL콜레스테롤 결측치값을 전체 LDL콜레스테롤 평균값으로 보간하기**

```python
health['LDL콜레스테롤']= health['LDL콜레스테롤'].fillna(health['LDL콜레스테롤'].mean())
```

단, 결측치가 많이 존재하는 연령대일 경우 해당 결측치 값을 전체 LDL콜레스테롤 평균값으로 보간했기 때문에 추후에 구할 연령별콜레스테롤의 값이 상대적으로 커진다는 한계가 존재한다.

**연령별로 LDL콜레스테롤 평균값 구해서 health 데이터에 컬럼으로 추가하기**

```python
health['연령별콜레스테롤']=health.groupby(['Age Name'])['LDL콜레스테롤'].transform('mean')
display(health)
```

**비만도(체중/키의제곱) 변수 추가**

```python
health['BMI']=(health['weight']/health['height']**2)*10000
display(health)
```
